```python
# Import all required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df['joining_date'] = pd.to_datetime(df['joining_date'], format = '%Y-%m-%d')
```

```python
df.dtypes
```

```python
df = df.rename(columns = {
    'emp_first_name' : 'first_name',
    'emp_last_name' : 'last_name'})
```

```python
df.columns = df.columns.str.title()
df.columns
```

```python
df['Remarks']
```

```python
df['Remarks'].value_counts()
```

```python
df['Remarks'] = df['Remarks'].str.strip()
```

```python
df['Remarks'] = df['Remarks'].str.title()
```

```python
df.shape
```

```python
df = df.drop_duplicates()
```

```python
df.isnull().sum()
```

```python
mean_salary = df['Salary'].mean()
mean_salary
df = df.copy()
df['Salary'] = df['Salary'].fillna(mean_salary)
```

```python
df.columns
```

```python
df['Remarks']
```

```python
df['Remarks'].value_counts()
```

```python
df['Remarks'] = df['Remarks'].fillna(df['Remarks'].mode()[0])
```

```python
plt.boxplot(df['Performance_Score'])
```

```python
plt.boxplot(df['Salary'])
```

```python
Q1 = df['Salary'].quantile(0.25)
Q3 = df['Salary'].quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
Lower_bound = Q1 - 1.5 * IQR
Upper_bound = Q3 + 1.5 * IQR
Lower_bound
Upper_bound
```

```python
df = df[(df['Salary'] >= Lower_bound) & (df['Salary'] <= Upper_bound)]
```

```python
plt.boxplot(df['Salary'])
```

```python
#1.WHAT IS TH EDISTRIBUTION OF EMPLOYEE SALARIES
```

```python
sns.set_theme(style = 'whitegrid' , palette = 'Set2')
plt.figure(figsize = (8,5))
sns.histplot(df['Salary'], kde = True , color = 'teal')
plt.title('Salary Distribution')
plt.xlabel('Salary Distribution')
plt.ylabel('No. of Employees')
plt.show()
```

```python
df.columns
```

```python
#2.HOW MANY EMPLOYEES ARE IN EACH DEPARTMENT
```

```python
plt.figure(figsize = (7,5))
sns.countplot(x = 'Department', data = df, palette = 'Set2')
plt.title('Employees Count Per Department')
plt.xlabel('Employee Department')
plt.ylabel('Count of Employees')
```

```python
#3.WHAT IS THE AVERAGE SALARY BY DEPARTMENT
```

```python
df.columns
```

```python
plt.figure(figsize = (7,5))
sns.barplot(x = 'Department', y = 'Salary', data = df, estimator = np.mean, palette = 'pastel')
plt.title = ('Average Salary by Department')
```
